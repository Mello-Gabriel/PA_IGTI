```python
texto=open('vale.csv').read()
```

```python
#Definindo a lista de stopwords
stopwords= set(STOPWORDS)

#Adicionando a lista stopwords em portuguÃªs
new_words = []
with open("stopwords.txt", 'r', encoding='utf-8') as f:
    [new_words.append(word) for line in f for word in line.split()]

new_stopwords = stopwords.union(new_words)
#new_stopwords
```

```python
mascara=np.array(Image.open('vale.png'))
```

```python
wc=WordCloud(background_color='Black',stopwords=new_stopwords, mask=mascara, min_font_size=3).generate(texto)
```

```python
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
```
